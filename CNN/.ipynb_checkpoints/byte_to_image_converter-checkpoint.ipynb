```python
import os, sys, time
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from keras.preprocessing import image

sys.path.insert(0, os.path.abspath('./utils/'))
import util as utils
%matplotlib inline
```

```python
bytecodes_dir = '../dataset_bytes/'
```

```python
bytecodes_files = [ os.path.splitext(os.path.basename(file))[0] for file in sorted(os.listdir(bytecodes_dir)) ]
```

```python
bytecodes_files
```

```python
dataset_dir = dataset_dir = os.path.abspath('./dataset/')

train_dir = os.path.join(dataset_dir, 'train')
test_dir = os.path.join(dataset_dir, 'test')
```

```python
if not os.path.exists(dataset_dir):
    os.mkdir(dataset_dir)

if not os.path.exists(os.path.join(dataset_dir, train_dir)):
    os.mkdir(os.path.join(dataset_dir, train_dir))

if not os.path.exists(os.path.join(dataset_dir, test_dir)):
    os.mkdir(os.path.join(dataset_dir, test_dir))
```

```python
image_set = []

start_convert_images = time.monotonic()

for filename in bytecodes_files:
    image_set.append([ utils.byte2img(os.path.join(bytecodes_dir, filename+'.bytes')), filename ])

end_convert_images = time.monotonic()
```

```python
print(end_convert_images - start_convert_images)
```

```python
len(image_set)
```

```python
for image in image_set:
    image[0].save(os.path.join(dataset_dir, image[1]+'.png'))
```

```python
from glob import glob
```

```python
image_files = sorted(glob(dataset_dir+'/*.png'))
```

```python
training_set_size = round(len(image_files) * 0.7)
```

```python
image_files_training = image_files[:training_set_size]
```

```python
image_files_testing = image_files[training_set_size:]
```

```python
for image_file in image_files_training:
    os.rename(image_file, os.path.join(train_dir, os.path.basename(image_file)))

for image_file in image_files_testing:
    os.rename(image_file, os.path.join(test_dir, os.path.basename(image_file)))
```
